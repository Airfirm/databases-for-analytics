```python
## Import necessary libraries

import pandas as pd
from sqlalchemy import create_engine
```

```python
file_path = r"C:\Users\ogsal\OneDrive\Desktop\NWMSU\exercise_07\title.episode.tsv"

df_episode = pd.read_csv(
    file_path,
    sep="\t",
    encoding="cp1252",
    encoding_errors="ignore",
    na_values="\\N",
    low_memory=False
)

df_episode.head()

```

```python
df_episode = df_episode.rename(columns={
    "parentTconst": "parent_tconst",
    "seasonNumber": "season_number",
    "episodeNumber": "episode_number"
})

df_episode["season_number"] = pd.to_numeric(df_episode["season_number"], errors="coerce")
df_episode["episode_number"] = pd.to_numeric(df_episode["episode_number"], errors="coerce")
```

```python
## Load into PostgreSQL (chunked & fast)

engine = create_engine(
    "postgresql+psycopg2://postgres:IreKayin01$$$@localhost:5432/movies_db"
)

df_episode.to_sql(
    "title_episode",
    engine,
    if_exists="append",
    index=False,
    chunksize=10000,
    method="multi"
)
```
